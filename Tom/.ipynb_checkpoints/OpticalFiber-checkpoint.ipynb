<h4>1 - Argmax Method</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set the plotting style to default with white backgrounds.
plt.style.use('default')
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'

# Read the data files and convert the columns to float.
input_pulse_data = pd.read_csv("../single_pulse_data_input.txt",
                               delim_whitespace=True,
                               header=None,
                               names=["Time", "Amplitude"]).astype(float).to_numpy()

output_pulse_data = pd.read_csv("../single_pulse_data_output.txt",
                                delim_whitespace=True,
                                header=None,
                                names=["Time", "Amplitude"]).astype(float).to_numpy()

# Define the desired x-axis limits (in seconds).
x_min = 0
x_max = 2e-8

# Create subplots that share the same x-axis.
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot the input pulse data.
axs[0].plot(input_pulse_data[:, 0], input_pulse_data[:, 1], label="Input Pulse", color='b')
axs[0].set_title('Input Pulse Data', color='black')
axs[0].set_ylabel('Amplitude', color='black')
axs[0].grid(True)
axs[0].legend()
axs[0].tick_params(axis='both', colors='black')
axs[0].set_xlim(x_min, x_max)

# Plot the output pulse data.
axs[1].plot(output_pulse_data[:, 0], output_pulse_data[:, 1], label="Output Pulse", color='r')
axs[1].set_title('Output Pulse Data', color='black')
axs[1].set_xlabel('Time', color='black')
axs[1].set_ylabel('Amplitude', color='black')
axs[1].grid(True)
axs[1].legend()
axs[1].tick_params(axis='both', colors='black')
axs[1].set_xlim(x_min, x_max)

# Identify the peak times using argmax.
input_peak_index = np.argmax(input_pulse_data[:, 1])
output_peak_index = np.argmax(output_pulse_data[:, 1])
input_peak_time = input_pulse_data[input_peak_index, 0]
output_peak_time = output_pulse_data[output_peak_index, 0]

# Print the peak times in units of 1e-8 seconds.
print(f"Input Peak index: {input_peak_index}")
print(f"Input Peak Time: {input_peak_time/1e-8:.3f} x 1e-8 s")
print(f"Output Peak index: {output_peak_index}")
print(f"Output Peak Time: {output_peak_time/1e-8:.3f} x 1e-8 s")

time_shift = output_peak_time - input_peak_time
time_shift_ns = time_shift * 1e9
print(f"Estimated time shift (output - input): {time_shift_ns:.3f} ns")

# Add vertical markers at the peak times.
axs[0].axvline(x=input_peak_time, color='k', linestyle='--', label='Input Peak')
axs[1].axvline(x=output_peak_time, color='k', linestyle='--', label='Output Peak')

# Optionally update legends to include the vertical lines.
axs[0].legend()
axs[1].legend()

plt.tight_layout()  # Adjust layout to prevent clipping.
plt.show()

if time_shift < 0:
    print("Note: A negative time shift indicates that the peak of the output pulse occurs "
          "before the input pulse's peak. Please verify the time references and data alignment.")
```

As we can see above, Chromatic Dispersion is taking place, as the light pulse is spread out. This makes obtaining the propagation delay difficult, as it is wavelength dependent

<h4>2 - Median + Argmax Method</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set the plotting style to default with white backgrounds.
plt.style.use('default')
plt.rcParams['axes.facecolor']   = 'white'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'

# Read the data files and convert the columns to float.
input_pulse_data = pd.read_csv("../single_pulse_data_input.txt",
                                 delim_whitespace=True,
                                 header=None,
                                 names=["Time", "Amplitude"]).astype(float).to_numpy()

output_pulse_data = pd.read_csv("../single_pulse_data_output.txt",
                                  delim_whitespace=True,
                                  header=None,
                                  names=["Time", "Amplitude"]).astype(float).to_numpy()

# Define the desired x-axis limits (in seconds).
x_min = 0
x_max = 2e-8

# Create subplots that share the same x-axis.
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot the input pulse data.
axs[0].plot(input_pulse_data[:, 0], input_pulse_data[:, 1],
            label="Input Pulse", color='b')
axs[0].set_title('Input Pulse Data', color='black')
axs[0].set_ylabel('Amplitude', color='black')
axs[0].grid(True)
axs[0].tick_params(axis='both', colors='black')
axs[0].set_xlim(x_min, x_max)

# Plot the output pulse data.
axs[1].plot(output_pulse_data[:, 0], output_pulse_data[:, 1],
            label="Output Pulse", color='r')
axs[1].set_title('Output Pulse Data', color='black')
axs[1].set_xlabel('Time', color='black')
axs[1].set_ylabel('Amplitude', color='black')
axs[1].grid(True)
axs[1].tick_params(axis='both', colors='black')
axs[1].set_xlim(x_min, x_max)

# --- Compute the input time using argmax ---
input_peak_index = np.argmax(input_pulse_data[:, 1])
input_peak_time = input_pulse_data[input_peak_index, 0]

# --- Compute the median time of the output pulse ---
cumulative = np.cumsum(output_pulse_data[:, 1])
normalized_cumulative = cumulative / cumulative[-1]
median_time = np.interp(0.5, normalized_cumulative, output_pulse_data[:, 0])

# Convert times to units of 1e-8 seconds for printing.
input_time_units = input_peak_time / 1e-8
median_time_units = median_time / 1e-8

print(f"Input (argmax) Time: {input_time_units:.3f} x 1e-8 s")
print(f"Output Median Time: {median_time_units:.3f} x 1e-8 s")

time_shift = median_time - input_peak_time
time_shift_ns = time_shift * 1e9
print(f"Estimated time shift (output - input) using median: {time_shift_ns:.3f} ns")

# --- Add vertical markers to the plots ---
axs[0].axvline(x=input_peak_time, color='k', linestyle='--', label="Input Peak")
axs[1].axvline(x=median_time, color='k', linestyle='--', label="Output Median")

# Update legends to include the vertical markers.
axs[0].legend()
axs[1].legend()

plt.tight_layout()  # Adjust layout to prevent clipping.
plt.show()
```

<h4>3 - Mean + Argmax Method</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set the plotting style to default with white backgrounds.
plt.style.use('default')
plt.rcParams['axes.facecolor']   = 'white'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'

# Read the data files and convert the columns to float.
input_pulse_data = pd.read_csv("../single_pulse_data_input.txt",
                                 delim_whitespace=True,
                                 header=None,
                                 names=["Time", "Amplitude"]).astype(float).to_numpy()

output_pulse_data = pd.read_csv("../single_pulse_data_output.txt",
                                  delim_whitespace=True,
                                  header=None,
                                  names=["Time", "Amplitude"]).astype(float).to_numpy()

# Define the desired x-axis limits (in seconds).
x_min = 0
x_max = 2e-8

# Create subplots that share the same x-axis.
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot the input pulse data.
axs[0].plot(input_pulse_data[:, 0], input_pulse_data[:, 1],
            label="Input Pulse", color='b')
axs[0].set_title('Input Pulse Data', color='black')
axs[0].set_ylabel('Amplitude', color='black')
axs[0].grid(True)
axs[0].tick_params(axis='both', colors='black')
axs[0].set_xlim(x_min, x_max)

# Plot the output pulse data.
axs[1].plot(output_pulse_data[:, 0], output_pulse_data[:, 1],
            label="Output Pulse", color='r')
axs[1].set_title('Output Pulse Data', color='black')
axs[1].set_xlabel('Time', color='black')
axs[1].set_ylabel('Amplitude', color='black')
axs[1].grid(True)
axs[1].tick_params(axis='both', colors='black')
axs[1].set_xlim(x_min, x_max)

# --- Compute the input time using argmax ---
input_peak_index = np.argmax(input_pulse_data[:, 1])
input_peak_time = input_pulse_data[input_peak_index, 0]

# --- Compute the centroid (weighted mean) time of the output pulse ---
output_centroid_time = np.sum(output_pulse_data[:, 0] * output_pulse_data[:, 1]) / np.sum(output_pulse_data[:, 1])

# Convert times to units of 1e-8 seconds for printing.
input_time_units = input_peak_time / 1e-8
centroid_time_units = output_centroid_time / 1e-8

print(f"Input (argmax) Time: {input_time_units:.3f} x 1e-8 s")
print(f"Output Centroid Time: {centroid_time_units:.3f} x 1e-8 s")

# Compute the time shift (output minus input) and convert to nanoseconds.
time_shift = output_centroid_time - input_peak_time
time_shift_ns = time_shift * 1e9
print(f"Estimated time shift (output - input) using centroid: {time_shift_ns:.3f} ns")

# --- Add vertical markers to the plots ---
# For the input plot, mark the argmax (input peak) time.
axs[0].axvline(x=input_peak_time, color='k', linestyle='--', label="Input Peak")
# For the output plot, mark the centroid (weighted mean) time.
axs[1].axvline(x=output_centroid_time, color='k', linestyle='--', label="Output Centroid")

# Update legends to include the vertical markers.
axs[0].legend()
axs[1].legend()

plt.tight_layout()  # Adjust layout to prevent clipping.
plt.show()
```

<h4> 4 - Arbitary placement method </h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set the plotting style to default with white backgrounds.
plt.style.use('default')
plt.rcParams['axes.facecolor']   = 'white'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'

# Read the output pulse data and convert to float.
output_pulse_data = pd.read_csv("../single_pulse_data_output.txt",
                               
                                  header=None,
                                  names=["Time", "Amplitude"]).astype(float).to_numpy()

# Define the desired x-axis limits (in seconds).
x_min = 0
x_max = 2e-8 

# Define a user-specified marker time (in seconds).
# Change this value to set the vertical marker at a different location.
time_shift = 2.13e-9
marker_time = time_shift + (0.1e-8)


# Create the plot for the output pulse.
plt.figure(figsize=(10, 6))
plt.plot(output_pulse_data[:, 0], output_pulse_data[:, 1],
         label="Output Pulse", color='r')
plt.title("Output Pulse Data", color='black')
plt.xlabel("Time (s)", color='black')
plt.ylabel("Amplitude", color='black')
plt.grid(True)
plt.xlim(x_min, x_max)

# Add a vertical dashed line at the user-defined marker time.
plt.axvline(x=marker_time, color='k', linestyle='--',
            label=f"Marker at {marker_time:.2e} s")

plt.legend()
plt.tight_layout()
plt.show()
```
